```python
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original
```

```python
median_salary = df_original['salary_year_avg'].median()

median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)
```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original.loc[:5, 'salary_year_avg']
```

```python
#so we can see that by altering altered df only. original one also got changed
# it means df_altered = df_original will not create a exact copy

# check by using id
id(df_original) == id(df_altered)
```

```python
# so use copy
df_original = dataset['train'].to_pandas()
df_altered = df_original.copy()

id(df_altered) == id(df_original)
```
